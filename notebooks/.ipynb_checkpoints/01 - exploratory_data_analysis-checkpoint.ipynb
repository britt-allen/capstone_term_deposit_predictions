```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
#importing the usual suspects (packages/libraies)

import warnings
warnings.simplefilter(action = 'ignore')
#preventing those ugly pink warnings from printing out

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
#displayng my plots inline and changing the inline backeng default figure format so it's cleaner
```

```python
bmd_df = pd.read_csv('./bank.csv')

#reading in my DataFrame and assigning it the variable 'bmd_df' â€“ which stands for bank marketing DataFrame
```

```python
bmd_df = bmd_df.drop(columns = 'duration')
```

Dropped the above feature as it is only known once our target (y) is known and my intention is to have a realistic predictive model.

```python
bmd_df['target'] = bmd_df.deposit.map(lambda x: 1 if x == 'yes' else 0)

#converting my target variable into '1s' and '0s' for my binary classification algorithms
```

```python
b = [18, 35, 50, 72, 90, np.inf]
names = ['Millennials (18-34)', 
         'Gen X (35-49)',
         'Baby Boomers (50-71)',
         'Silent Gen (72-89)',
         'GI Gen (90+)']

bmd_df['age_group'] = pd.cut(x = bmd_df.age, bins = b, labels = names, right = False)
```

I engineered a new feature above, <b>age_group</b>, to bucket customers by their generational age groups as I think it will make data visualization and analysis easier and more meaningful.

```python
data = {'Feature': ['age', 'job', 'marital', 'education', 'default', 'balance', 'housing',
                  'loan', 'contact', 'day', 'month', 'campaign', 'pdays', 'previous',
                  'poutcome', 'deposit', 'target', 'age_group'],
        'Type': ['int64', 'object', 'object', 'object', 'object', 'int64', 'object', 'object', 'object',
                 'int64', 'object', 'int64', 'int64', 'int64', 'object', 'object', 'int64', 'category'], 
        'Description': ["numeric age of the customer",
                        "type of job (categorical: admin, technician, services, management, retired, blue-collar, unemployed, \nentrepreneur, housemaid, unknown, self-employed, student)",
                        "marital status (categorical: divorced (divorced or widowed), married, single)",
                        "level of education (categorical: primary (pre-high school), secondary (high school), \ntertiary (university), unknown)", 
                        "has credit in default? (binary: no, yes)",
                        "balance of the customer (numeric)",
                        "has housing loan? (binary: no, yes)",
                        "has personal loan? (binary: no, yes)", 
                        "communication device type (categorical: cellular, telephone, unknown)", 
                        "last contact day of the month (categorical: 1, 2, ..., 30, 31)",
                        "last contact month of year (categorical: jan, feb, ..., nov, dec)", 
                        "number of contacts performed during this campaign and for this client (numeric; includes last contact)", 
                        "number of days that passed by after the client was last contacted from a previous campaign \n(numeric; -1 means client was not previously contacted)",
                        "number of contacts performed before this campaign and for this client (numeric)",
                        "outcome of the previous marketing campaign (categorical: failure, other, success, unknown)",
                        "has the client subscribed a term deposit? (binary: yes,no)",
                        "has the client subscribed a term deposit? (binary, numeric: 1 = yes, 0 = no)",
                        "generational age groupings of the customer (categorical: Millennials (18-34), Gen X (35-49), \nBaby Boomers (50-71), Silent Gen (72-89), GI Gen (90+))"]}

# creating a variable "data" that will house a dictionary with 3 key:value pairs to be fed into a DataFrame below
```

```python
data_dict = pd.DataFrame(data = data)
data_dict

#created a DataFrame, data_dict, of information regarding all features that will be fed into the table below
```

```python
plt.figure(figsize=(20, 20))
#setting the figure size

table = plt.table(cellText=np.array(data_dict),
                  colWidths = [0.10, 0.10, 0.80],
                  colLabels= data_dict.columns,
                  colLoc = 'left',
                  cellLoc= 'left',
                  loc='center',
                  bbox=[0,0,1,1])
#creating the table

def set_pad_for_column(col, pad=0.1):
    cells = [key for key in table._cells if key[1] == col]
    for cell in cells:
        table._cells[cell].PAD = pad
set_pad_for_column(col=2, pad=0.015)
#scaled down the padding of the "Description" column which initially had a lot of unecessary padding 

for (row, col), cell in table.get_celld().items():
    if (row == 0) or (col == 0):
        cell.set_text_props(fontproperties=FontProperties(weight='heavy'))
#code that will allow for the first row and column to be bolded

table.auto_set_font_size(False) #making it so that the font size is not automatically set
table.set_fontsize(20)          #manually setting the table font size
table.scale(1, 3)               #changing the scale of my table to be the size I want
plt.axis('off')                 #turning off the plot axis as it make the table unreable when left "on"
plt.tight_layout()              #making my image fit within the limits of the figure to avoid clipping
plt.savefig("data_dict.png")    #choosing image name & file type and saving it locally
plt.show();                     #displaying my figure inline below
```

Above, I created a table leveraging matplotlib to display my data dictionary in a cleaner format. I made stylistic edits to control aspects such as padding, alignment, font size, bolding, etc. I saved the figure to my local repo for use in my presentation.

# Exploratory Data Analysis

```python
summary_stats = pd.DataFrame(data=round(bmd_df.describe(), 0), dtype=int)
summary_stats

#printing out summary stats for my numerical features
```

```python
fig, ax = plt.subplots(figsize = (10, 6))
#specifying my figure size

plot = plt.bar(bmd_df.deposit.unique(), bmd_df.deposit.value_counts(), color='c')
#visualizing my 'deposit' feature with a bar chart

plt.xticks(fontsize=17, rotation=0)
plt.yticks(fontsize=17)
plt.ylabel("count", fontsize=20, labelpad=15)
plt.title(f'Made a Term {bmd_df.deposit.name.capitalize()}?', fontsize=30)
#above I created my bar chart, set my xticks font size & rotation, and set my yticks font size
#also set my ylabel and plot title

rects = ax.patches
labels = [i for i in round(bmd_df.deposit.value_counts(normalize=True),2)*100]

for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height + 5, f'{int(label)}%',
            ha='center', va='bottom', fontsize=15)
#created a for loop to assign value labels to the top of my bars

plt.ylim(0, 6500)                   #expanding the limits of my y-axis to make my image more readable
plt.tight_layout()                  #making my image fit within the limits of the figure to avoid clipping
plt.savefig("target_bar_chart.png") #choosing image name & file type and saving it locally
plt.show();                         #displaying my figure inline below
```

```python
fig, ax = plt.subplots(figsize = (11, 8))
#specifying my figure size

sns.pointplot(x = bmd_df.deposit, y = bmd_df.balance, hue = bmd_df.age_group, kind='point', ci=None)
plt.title(f'Mean Balances by Age Group & Term Deposit', fontsize=26.5)
#visualizing the realtionship between average balances by deposit and by age
#creating my plot title

plt.tight_layout()      #making my image fit within the limits of the figure to avoid clipping
plt.savefig("deposit_balance_agegroup-hue_mean_pointplot.png") #choosing image name & file type and saving it locally
plt.show();             #displaying my figure inline below
```

```python
fig, ax = plt.subplots(figsize = (11, 8))
#specifying my figure size

ax = sns.countplot(x = bmd_df.age_group, order=['Gen X (35-49)', 'Millennials (18-34)', 'Baby Boomers (50-71)',
                                                'Silent Gen (72-89)', 'GI Gen (90+)'])
#visualizing counts and percentages of total for customers in each age grouping

plt.xticks(fontsize=16, rotation=10)
plt.yticks(fontsize=16)
plt.ylabel("count", fontsize=20, labelpad=15)
plt.xlabel("age group", fontsize=20, labelpad=10)
plt.title('Age Group Percentages of Total', fontsize=30)
#above I set my xticks font size & rotation and set my yticks font size
#also set my ylabel and plot title

rects = ax.patches
labels = [i for i in round(bmd_df.age_group.value_counts(normalize=True),2)*100]

for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height + 5, f'{int(label)}%',
            ha='center', va='bottom', fontsize=15)
#created a for loop to assign value labels to the top of my bars    

plt.ylim(0, 5000)
plt.tight_layout()   #making sure my layout is tight to avoid clipping when saving my figure
plt.savefig("agegroup_percentages_barplot.png") #choosing image name & file type and saving it locally
plt.show();          #displaying my figure inline below
```

```python
fig, ax = plt.subplots(figsize = (11, 6))

plot = plt.bar(bmd_df.month.unique(), bmd_df.month.value_counts(), color='c')
plt.xticks(fontsize=17, rotation=0)
plt.yticks(fontsize=17)
plt.ylabel("count", fontsize=20, labelpad=15)
plt.title(f'Marketing Activity by {bmd_df.month.name.capitalize()}', fontsize=26.5)

rects = ax.patches
labels = [i for i in round(bmd_df.month.value_counts(normalize=True),2)*100]

for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height + 5, f'{int(label)}%',
            ha='center', va='bottom', fontsize=15)
    
plt.ylim(0, 3200)
plt.savefig("month_conv_bar_chart.png") #choosing image name & file type and saving it locally
plt.tight_layout()
plt.show();
```

```python
d = {'May': [f'{int(round(925 / (925+1899), 3)*100)}%'], 'Jan': [f'{float(round(142 / (142+202), 3)*100)}%'],
     'Jul': [f'{float(round(627 / (627+887), 3)*100)}%'], 'Nov': [f'{int(round(403 / (403+540), 3)*100)}%'],
     'Jun': [f'{int(round(546 / (546+676), 3)*100)}%'], 'Aug': [f'{int(round(688 / (688+831), 3)*100)}%'],
     'Feb': [f'{int(round(441 / (441+335), 3)*100)}%'], 'Apr': [f'{int(round(577 / (577+346), 3)*100)}%'],
     'Oct': [f'{int(round(323 / (323+69), 3)*100)}%'], 'Sep': [f'{int(round(269 / (269+50), 3)*100)}%'],
     'Mar': [f'{int(round(248 / (248+28), 3)*100)}%'], 'Dec': [f'{int(round(100 / (100+10), 3)*100)}%']}
```

```python
month_conv = pd.DataFrame(data=d)
month_conv

#created a DataFrame, month_conv, of months and corresponding conversion rates to be fed into the table below
```

```python
fig, ax = plt.subplots(figsize = (14, 1.75))
#specifying my figure size

table_months = plt.table(cellText=np.array(month_conv),
                  colLabels= month_conv.columns,
                  colLoc = 'center',
                  cellLoc= 'center',
                  loc='center',
                  bbox=[0,0.2,.65,.65])
#creating the table

plt.title(f'Conversion Rates (CV) by {bmd_df.month.name.capitalize()}', fontsize=22, loc='center')
#setting my plot title

for (row, col), cell in table_months.get_celld().items():
    if (row == 0):
        cell.set_text_props(fontproperties=FontProperties(weight='heavy'))
#code that will allow for the header row to be bolded
        
table.auto_set_font_size(False)     #making it so that the font size is not automatically set
table.set_fontsize(24)              #manually setting the table font size
table.scale(2.5, 2.5)               #changing the scale of my table to be the size I want
plt.axis('off')                     #turning off the plot axis as it make the table unreable when left "on"
plt.subplots_adjust(top=2)          #adjusting the plotting space a but more
plt.tight_layout()                  #making sure my layout is tight to avoid clipping when saving my figure
plt.savefig("month_conv_table.png") #choosing image name & file type and saving it locally
plt.show();                         #displaying my figure inline below
```
